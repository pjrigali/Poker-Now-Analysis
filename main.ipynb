```python
import time
import re

main_string = "Lorem ipsum dolor sit amet, consectetur adipiscing elit."
substring = "consectetur"

# Test the 'in' keyword approach
start_time = time.time()
for _ in range(1000000):
    if substring in main_string:
        pass
in_duration = time.time() - start_time

# Test the 'str.find()' method approach
start_time = time.time()
for _ in range(1000000):
    if main_string.find(substring) != -1:
        pass
find_duration = time.time() - start_time

start_time = time.time()
for _ in range(1000000):
    if main_string.startswith('Lorem'):
        pass
end_duration = time.time() - start_time

start_time = time.time()
for _ in range(1000000):
    if re.search(substring, main_string):
        pass
re_duration = time.time() - start_time

print("'in' keyword approach duration:", in_duration)
print("'str.find()' method approach duration:", find_duration)
print("'str.endswith()' method approach duration:", end_duration)
print("'str.re()' method approach duration:", re_duration)
```

```python
from poker.utils.tools import percent
from poker.utils.tools import native_mean, native_std, native_max, native_median, native_variance
import pandas as pd


names = ('Flynn', 'Pfanz', 'Peter', 'Kaz', 'Brian', 'Henry', 'Robert', 'Robby')
sep = 10
cont = {(round((i / sep) - (1.0 / sep - .01), 3), i / sep): [] for i in list(range(1, sep + 1))}
itm = 'player_perc_current'
final = {}

for n in names:
    dct = {p.rows[0]['game_id']: {'start': p.rows[0]['hand_startTs'], 'end': None, 'moves': []}}
    for i in p.rows:
        if i.get('playerId') and i['playerId'] in grouped[n] and not i.get('allIn'):
            if i['game_id'] not in dct:
                dct[i['game_id']] = {'start': i['hand_startTs'], 'end': None, 'moves': []}
            else:
                # if i.get('move') and i['move'] in ('Bet', 'Call', 'Raise', 'Uncalled Bet'):
                if i.get('move') and i['move'] in ('Bet'):
                    dct[i['game_id']]['moves'].append(i)
            
            if i['hand_endTs']:
                dct[i['game_id']]['end'] = i['hand_endTs']

    for k, v in dct.items():
        temp = []
        for i in v['moves']:
            temp.append({
                        #  'starting_total': sum(v1 for k1, v1 in i.items() if k1.startswith('starting_chips_')),
                        #  'player_perc_starting': percent(i['value'], i[f"starting_chips_{i['player']}"]),
                         'player_perc_current': percent(i['value'], i[f"current_chips_{i['player']}"]) if i.get(f"current_chips_{i['player']}") else percent(i['value'], i[f"starting_chips_{i['player']}"]),
                        'time_taken': (i['at'] - i['decisionTime']).total_seconds(),
                        'time_into_game': percent((i['at'] - v['start']).total_seconds(), (v['end'] - v['start']).total_seconds()),
                        'perc_pot': percent(i['value'], i['pot'] - i['value']),
                        'value': i['value'],
                        # 'pot': i['pot'] - i['value']
                        })
        dct[k]['moves'] = temp
    
    cont2 = dict(cont)
    for k, v in dct.items():
        for i in v['moves']:
            for j in cont2:
                if i['time_into_game'] >= j[0] and i['time_into_game'] <= j[1] and i[itm]:
                    cont2[j].append(i[itm])
    # cont3 = sum([v for k, v in cont2.items()], [])
    cont2 = {k: round(native_mean(v), 3) for k, v in cont2.items()}
    cont3 = list(cont2.values())
    cont3 = {'': '', 
             'mean': native_mean(cont3),
             'median': native_median(cont3), 
             'std': native_std(cont3), 
             'variance': native_variance(cont3), 
            #  'max': native_max(cont3)
             }
    final[n] = {**cont2, **cont3}

final = pd.DataFrame(final, columns=names)
```

```python
from poker.utils.tools import calculate_hand

lst = []
for i in p.hands:
    if i.mycards:
        d = {'my_cards': i.mycards,
             'table_cards': sum([i.table_cards[j] for j in ('flop', 'turn', 'river') if i.table_cards.get(j)], []),
             'hand': None}
        d['hand'] = calculate_hand(d['my_cards'] + d['table_cards'])
        lst.append(d)

tot, better, worse, b_cards, w_cards = 0, 0, 0, [], []
for i in lst:
    tot += 1
    if i['hand'].endswith('High') or i['hand'] == 'Pair':
        worse += 1
        w_cards.extend(i['my_cards'])
    else:
        better += 1
        b_cards.extend(i['my_cards'])

b_dct, w_dct = {}, {}
for i in b_cards:
    if i not in b_dct:
        b_dct[i] = 0
    b_dct[i] += 1

for i in w_cards:
    if i not in w_dct:
        w_dct[i] = 0
    w_dct[i] += 1
    
print(f"better than a two pair: ({round(better / tot, 3)})")
print(f"worse than a two pair: ({round(worse / tot, 3)})")

idx = sorted(list(set(list(b_dct.keys()) + list(w_dct.keys()))))
df = pd.DataFrame(index=idx)
w = sorted([(k, v) for k, v in w_dct.items()])
b = sorted([(k, v) for k, v in b_dct.items()])
df['worse'] = [i[1] for i in w]
df['better'] = [i[1] for i in b]
df = df.sort_values(by=['better'], ascending=False)
```

```python
from poker.utils.functions import get_player_id

def group_results(grouped: dict, rows: list) -> dict:
    g, g_names = {}, {}
    for k, v in grouped.items():
        for i in v:
            g_names[i], g[i] = k, []
    
    for i in rows:
        if i['player']:
            p = get_player_id(i['player'])
            if p:
                if p not in g:
                    g[p] = []
            
            if i['result']:
                if isinstance(i['result'], (list, tuple)):
                    for j in i['result']:
                        g[p].append(i['result'])
                else:
                    g[p].append(i['result'])
                    
    d = {i: [] for i in grouped}
    for k, v in g.items():
        if not g_names.get(k):
            d[k] = []
                  
        if k in g_names:
            d[g_names[k]].extend(v)
        else:
            d[k].extend(v)
            
    return d           

    
```
